```python
from bs4 import BeautifulSoup
from bs4 import Comment
from lxml import html
import json
import csv
import requests
import re
import time
```

#### Extract all links available

```python
indexlink = 'https://reeis.usda.gov/web/crisprojectpages/'
directory = requests.get(indexlink).text
```

```python
soup = BeautifulSoup(directory, 'lxml')
```

```python
links = []
for link in soup.find_all('a'):
    linkfound = link.get('href')
    if re.match('\d+-.+\.html', linkfound):
        links.append(linkfound)
```

```python
try:
    with open('failed.txt', 'r') as file:
        memoria = json.loads(file.read())
except Exception:
    memoria = []
```

#### For each link, grab info

```python
try:
    with open('failed.txt', 'r') as file:
        memoria = json.loads(file.read())
except Exception:
    memoria = []

with open('successful.csv','a') as success:
    csvwriter = csv.writer(success)
    for link in links[71974:]:
        try:
            time.sleep(1)
            projectpage = requests.get(indexlink+link).text
            psoup = BeautifulSoup(projectpage, 'lxml')

            ## Remove HTML comments <--- text --->
            comments = psoup.find_all(string=lambda text:isinstance(text,Comment))
            for x in comments:
                x.extract()

            ## Grab the first frame in HTML
            tags = []
            
            content1 = psoup.find('div', class_='container second')
            for div in content1.find_all('div'):
                txt = div.text.strip()
                if txt:
                    tags.append(txt)

            # Grab address up to non-tech summary       
            cleaner = {}
            
            content2 = psoup.find('div', class_='third old-table')
            content2 = content2.find_all('div')
            
            word = ''
            current = None
            
            cleaner['Title'] = psoup.find('div', { "id" : "title" }).text.strip()

            for div in content2[:6]:
                for tag in div:
                    if tag.name == 'b':
                        current = tag.string
                        word = ''
                    elif tag.string != None:
                        word += tag.string + ' '
                        if current != None:
                            cleaner[current] = word.strip()

            ## Combine strings from address
            cleaner['Recipient Organization'] = ' '.join(cleaner['Recipient Organization'].split())

            i = iter(tags)
            comp = dict(zip(i, i))
            comp.update(cleaner)

            print(comp['Accession No.'])
            
            text = json.dumps(comp)
            csvwriter.writerow([text])

        except Exception:
            memoria.append(link)
            print(link, 'failed. Try this one later')
            
            textmemoria = json.dumps(memoria)

            with open('failed.txt', 'w') as file:
                file.write(textmemoria)
```
